```python
#calcular e visualizar o spread do 90th salario do 10th salario
spread_col = clean_dado['Mid-Career 90th Percentile Salary'] - clean_dado['Mid-Career 10th Percentile Salary']
clean_dado.insert(1, 'Spread', spread_col)
clean_dado.head()
```

```python
#os 5 mais baixos Spreads
low_spread = clean_dado.sort_values('Spread')
low_spread[['Undergraduate Major', 'Spread']].head()
```

```python
#baseado na quest√£o: Using the .sort_values() method, can you find the degrees with the highest potential? Find the top 5 degrees with the highest values in the 90th percentile.
high_potential = clean_dado.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
high_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
#encontrar os 5 maiores Spreads
high_spread = clean_dado.sort_values('Spread', ascending=False)
high_spread[['Undergraduate Major', 'Spread']].head()
```

```python
#encontrar as maiores diferen√ßas salariais do inicio da carreira at√© o meio tempo de carreira
diference_salary = clean_dado['Mid-Career Median Salary'] - clean_dado['Starting Median Salary']
clean_dado.insert(1,'Diference Salary Between Starting and Mid', diference_salary)
clean_dado.head()
```

```python
high_salary = clean_dado.sort_values('Diference Salary Between Starting and Mid', ascending=False)
high_salary[['Undergraduate Major', 'Diference Salary Between Starting and Mid']].head()
```
